# Exploration

```python
from astra_core import CodeUnit, analyze_code_similarity

threshold = 0.8

units = [
    CodeUnit(
        id="student_a.py",
        content="""
def compute_sum(numbers):
    total = 0
    for n in numbers:
        total += n
    return total

data = [5, 10, 15, 20]
print(compute_sum(data))
""",
    ),
    CodeUnit(
        id="student_b.py",
        content="""
def sum_values(vals):
    acc = 0
    for v in vals:
        acc += v
    return acc

nums = [1, 2, 3, 4]
output = sum_values(nums)
print(output)
""",
    ),
]

result = analyze_code_similarity(
    units=units,
    threshold=threshold,
)

print(result)
```
